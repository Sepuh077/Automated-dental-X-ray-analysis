```python
train_images, train_pos_data, train_labels, test_images, test_pos_data, test_labels = load_single_tooth_data()

train_images.shape, train_pos_data.shape, train_labels.shape, test_images.shape, test_pos_data.shape, test_labels.shape
```

```python
net = SingleToothNet().to(device)
# summary(net, input_size=(400, 200));
# net = torch.load('../main_models/single_tooth_number_detection.pt')
```

```python
model = Model(
    model=net, 
    train_input=(train_images, train_pos_data), 
    train_label=train_labels,
    test_input=(test_images, test_pos_data),
    test_label=test_labels
)
```

```python
model.compile(
    optimizer=torch.optim.Adam, 
    lr=0.001,
    loss=torch.nn.CrossEntropyLoss()
)
```

```python
model.train(
    epochs=200,
    batch_size=64
)
```

```python
model.compile(
    optimizer=torch.optim.Adam, 
    lr=0.00001,
    loss=torch.nn.CrossEntropyLoss()
)
model.train(
    epochs=90,
    batch_size=16
)
```
